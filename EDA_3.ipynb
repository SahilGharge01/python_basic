```python
plt.figure(figsize=(15,15))
plt.suptitle('Univarient Analysis On Numerical Dataset', fontsize = 20, fontweight = 'bold')

for i in range(0, len(Numerical_columns)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x = df[Numerical_columns[i]], shade = True, color='r')
    plt.xlabel(Numerical_columns[i])
    plt.tight_layout()
```

# Categorical data analysis

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univarient Analysis On Categorical Dataset', fontsize = 20, fontweight = 'bold')
category = ['Type', 'Content Rating']
for i in range (0, len(category)):
    plt.subplot(2 , 2, i+1)
    sns.countplot(x = df[category[i]], palette = 'Set2')
    plt.xlabel(df[category[i]])
    plt.xticks(rotation = 45)
    plt.tight_layout()
```
