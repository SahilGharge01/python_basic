```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
```

```python
df = pd.read_csv('cleaned_google_play_store_data.csv')
```

 # which category  is having largest numbers of installation

```python
df_cat_installs = df.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
```

```python
df_cat_installs
```

```python
df['Installs'] = df_cat_installs['Installs']/100000000
```

```python
df_cat_installs.head(10)
```

```python
plt.figure(figsize=(15,16))
sns.set_context('talk')
sns.set_style('darkgrid')

ax = sns.barplot(x = 'Installs', y = 'Category', data = df_cat_installs)


ax.set_title('Most popular Category in play store', size = 20)
plt.show()
```
